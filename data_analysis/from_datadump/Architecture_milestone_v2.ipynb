```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Goal

This notebook visualizes some of the results from the Architecture milestone v2 milestone. In particular visualizes the results of the bilingual segments available, per language after the processing steps by Pangeanic are applied to the data produced by NTUA.
Reference https://docs.google.com/document/d/1fZp2_lEB9_XXgO9uTwYM86R6ZhRo04sThau3YUuaKec/edit

```python
data=pd.read_csv('bilingual_data_filtering_steps.csv')
#table obtained from architecture document
```

```python
data
```

```python
data = data.sort_values('Step 5', ascending=True)
```

```python
data=data.set_index("Language")
```

```python
data
```

```python
data.plot( y="Step 5", kind="bar", figsize=(15, 5));
```

```python
fig, ax1 = plt.subplots(figsize=(10,5))
font_dict = {'fontsize': 11, 'fontweight': 'heavy', 'verticalalignment':'top'}
# These are in unitless percentages of the figure size. (0,0 is bottom left)
left, bottom, width, height = [0.68, 0.30, 0.2, 0.2]
ax2 = fig.add_axes([left, bottom, width, height])

ax1.barh(y=data.index, width=data["Step 5"]);
ax1.set_ylabel('# metadata text segments',{'fontsize': 12})
ax2.barh(y=data.index[1:5], width=data["Step 5"][1:5])
ax2.set_ylabel('# metadata \n text segments',{'fontsize': 12});
```
